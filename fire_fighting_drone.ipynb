```python
from ultralytics import YOLO
import cv2

# Load the video
video = cv2.VideoCapture("video9.mp4")

# Load the pretrained YOLOv8 model
model = YOLO("best.pt")

while True:
    grabbed, frame = video.read()
    if not grabbed:
        break

    frame = cv2.resize(frame, (640, 360))

    # Run object detection
    results = model.track(source=frame, verbose=False, conf=0.6, persist=True)

    for r in results:
        for i in range(len(r.boxes.cls)):
            class_label = r.names[r.boxes.cls[i].item()]
            color = (0, 0, 255)
            if class_label == 'Smoke':
                color = (0, 255, 0)
            if class_label == 'Human':
                color = (255, 0, 0)
            x1, y1, x2, y2 = map(int, r.boxes.xyxy[i])
            mid_x = (x1 + x2) // 2
            mid_y = (y1 + y2) // 2

            text_pos = (y1 + y2) // 4

            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(frame, f"Type: {class_label}", (x1, text_pos), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 1, cv2.LINE_AA)

    cv2.imshow('YOLOv8 Tracking', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```
